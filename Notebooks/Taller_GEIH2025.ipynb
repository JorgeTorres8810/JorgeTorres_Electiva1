```python
def clasificar_edad(x):
    if 18 <= x < 24:
        return "18-24"
    elif 24 <= x < 30:
        return "24-30"
    elif 30 <= x < 40:
        return "30-40"
    elif 40 <= x < 50:
        return "40-50"
    elif x >= 60:
        return "60+"
    else:
        return "Otros"
```

```python
# Generamos una nueva columna que clasifica las personas según rangos de edad
df_cgsse["grupo_edad"] = df_cgsse["P6040"].apply(clasificar_edad)

# Integramos la información de los desempleados con su respectivo grupo etario
df_desocupados_edad = pd.merge(
    df_desocupados[df_desocupados["DSI"] == 1],
    df_cgsse[["DIRECTORIO", "SECUENCIA_P", "ORDEN", "grupo_edad"]],
    on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
    how="left"
)

# Total de personas sin empleo agrupadas por departamento y rango de edad
desocupados_edad = (
    df_desocupados_edad
    .groupby(["DPTO", "grupo_edad"])["FEX_C18"]
    .sum()
    .reset_index()
    .rename(columns={"FEX_C18": "personas_desempleadas"})
)

# Vinculamos la información de ocupados con los grupos etarios
df_ocupados_edad = pd.merge(
    df_ocupados,
    df_cgsse[["DIRECTORIO", "SECUENCIA_P", "ORDEN", "grupo_edad", "P3271"]],
    on=["DIRECTORIO", "SECUENCIA_P", "ORDEN"],
    how="left"
)

# Total de personas empleadas por departamento y rango de edad
ocupados_edad = (
    df_ocupados_edad
    .groupby(["DPTO", "grupo_edad"])["FEX_C18"]
    .sum()
    .reset_index()
    .rename(columns={"FEX_C18": "personas_ocupadas"})
)

# Construimos la PEA (ocupados + desempleados) y calculamos la tasa de desempleo
pea_edad = pd.merge(ocupados_edad, desocupados_edad, on=["DPTO", "grupo_edad"], how="outer").fillna(0)
pea_edad["PEA"] = pea_edad["personas_ocupadas"] + pea_edad["personas_desempleadas"]
pea_edad["tasa_desempleo(%)"] = (pea_edad["personas_desempleadas"] / pea_edad["PEA"]) * 100
```
