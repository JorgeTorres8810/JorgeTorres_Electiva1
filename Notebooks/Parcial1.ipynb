```python
# Se importan las librerias necesarias 
import pandas as pd
import numpy as np
```

```python
# Se crea objeto "emicron" para leer la base de datos
emicron = pd.read_csv("https://github.com/niconomist98/DataAnalyticsUQ/raw/refs/heads/main/Datos/EMICRON/emicron_identificacion.csv")
```

```python
# Previsualizando los encabezados de la tabla
emicron
```

Desarrollo de los puntos:

```python
# 1. Calcule la cantidad de micronegocios que hay en el departamento del Quindío encabezados por hombres y mujeres entre los 18 y los 28 años 
# Filtrar micronegocios en Quindío encabezados por personas entre 18 y 28 años
microneg_quindio = emicron.query("COD_DEPTO == 63 and 18 <= P241 <= 28")['F_EXP'].sum()

# Mostrar el resultado como número entero
print(f"Cantidad de micronegocios en el Quindío encabezados por personas con edades entre 18 y 28 años: {int(microneg_quindio)}")
```

```python
# 2.	Cual es la ciudad en la que existen mas micronegocios encabezados por hombres entre los 20 y los 30 años

# Filtrar micronegocios encabezados por hombres entre 20 y 30 años
micron_hombres = emicron.query("P35 == 1 and 20 <= P241 <= 30")

# Agrupar por ciudad (o área) y sumar el factor de expansión
micron_por_ciudad = micron_hombres.groupby('AREA')['F_EXP'].sum().sort_values(ascending=False)

# Mostrar la ciudad con mayor cantidad estimada de micronegocios
ciudad_top = micron_por_ciudad.idxmax()
cantidad_top = int(micron_por_ciudad.max())

print(f"La ciudad con más micronegocios encabezados por hombres entre 20 y 30 años es: {ciudad_top} ({cantidad_top} micronegocios)")
```

```python
# 3.	En el departamento del Tolima ¿Cuántos micronegocios existen encabezados por mujeres que no tienen una persona que les ayude en el micronegocio?

# Filtrar micronegocios en Tolima encabezados por mujeres sin ayudante
microneg_tolima = emicron.query("COD_DEPTO == 73 and P35 == 2 and P3031 == 2")['F_EXP'].sum()

# Mostrar el resultado como número entero
print(f"Cantidad de micronegocios en Tolima encabezados por mujeres sin ayudante: {int(microneg_tolima)}")
```

```python
# 4. Cuantos micronegocios hay en la ciudad de Bogotá encabezados por mujeres que estén en el sector de la industria manufacturera
# Filtrar micronegocios en Bogotá encabezados por mujeres en industria manufacturera
microneg_bogota = emicron.query("COD_DEPTO == 11 and P35 == 2 and GRUPOS4 == 2")['F_EXP'].sum()

# Mostrar el resultado como número entero
print(f"Cantidad de micronegocios en Bogotá encabezados por mujeres en industria manufacturera: {int(microneg_bogota)}")
```

```python
# 5. Formular una pregunta con minimo dos variables de interes personal 
# ¿Cuantos micronegocios en el departamento del Quindio se dedican a la industria manufacturera?
# ¿Que propoposion ocupuan del total de micrinegicios del departamento?

# Micronegocios en Quindío dedicados a la industria manufacturera
microneg_manufactura = emicron.query("COD_DEPTO == 63 and GRUPOS4 == 2")['F_EXP'].sum()

# Total de micronegocios en el departamento del Quindío
total_microneg_quindio = emicron.query("COD_DEPTO == 63")['F_EXP'].sum()

# Calcular proporción como porcentaje
proporcion_manufactura = (microneg_manufactura / total_microneg_quindio) * 100

print(f"Cantidad de micronegocios en Quindío dedicados a la industria manufacturera: {int(microneg_manufactura)}")
print(f"Proporción sobre el total de micronegocios del departamento: {proporcion_manufactura:.2f}%")
```
